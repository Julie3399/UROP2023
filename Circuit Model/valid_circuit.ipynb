```python
def add_connection(b, piece, y, x, port):
    """Update the connection of the piece at (x, y)

    Args:
        b (class): an instance of board
        piece (class): an instance of input piece
        y (float): y coordinate of the input piece
        x (float): x coordinate of the input piece
        port (str): indication of the type of connection, dual, inp, out, trigger, waves or hold
    """

    # Connect with all pieces already on the board with the appropriate possible ports
    for el in b.pos[y][x]:  # element will be composed of a list of tuples of pieces and ports
        # Check if the current element is the same as the piece being connected
        if el[0] == piece:
            pass
        # Handle dual port connections
        elif port == "dual":
            if el[1] == "inp":
                directional_connect(piece, el[0], 'out', 'inp')
            if el[1] == "inp2":
                directional_connect(piece, el[0], 'out', 'inp2')
            if el[1] == "out":
                directional_connect(piece, el[0], 'inp', 'out')
            if el[1] == "trigger":
                directional_connect(piece, el[0], 'out', 'trigger')
            if el[1] == "hold":
                directional_connect(piece, el[0], 'out', 'hold')
            if el[1] == "waves":
                directional_connect(piece, el[0], 'out', 'waves')
        # Handle other types of connections
        elif port == "inp":
            if el[1] == "out":
                directional_connect(piece, el[0], 'inp', 'out')
        elif port == "trigger":
            if el[1] == "out":
                directional_connect(piece, el[0], 'trigger', 'out')
        elif port == "waves":
            if el[1] == "out":
                directional_connect(piece, el[0], 'waves', 'out')
        elif port == "hold":
            if el[1] == "out":
                directional_connect(piece, el[0], 'hold', 'out')
        elif port == "out":
            if el[1] == "inp":
                directional_connect(piece, el[0], 'out', 'inp')
            if el[1] == "trigger":
                directional_connect(piece, el[0], 'out', 'trigger')
            if el[1] == "hold":
                directional_connect(piece, el[0], 'out', 'hold')
            if el[1] == "waves":
                directional_connect(piece, el[0], 'out', 'waves')
        elif port == "inp2":
            if el[1] == "out":
                directional_connect(piece, el[0], 'inp2', 'out')

    # Add the piece to the board so future connections can be made to it
    if port == "dual":
        i = (piece, "inp")
        b.pos[y][x].append(i)
        o = (piece, "out")
        b.pos[y][x].append(o)
    else:
        t = (piece, port)
        b.pos[y][x].append(t)
```
